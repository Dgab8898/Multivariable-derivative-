```python
import math
import sympy as sym
from IPython.display import display,Math
```

```python
from sympy.abc import x,y
f = 4*x**3*y**2 + 2*x**2*y**2 + x*y + 3*x

display(Math('\\frac{\\partial x}{\\partial x} = %s' %sym.latex(sym.diff(f,x))))
display(Math('\\frac{\\partial x}{\\partial y} = %s' %sym.latex(sym.diff(f,x))))
```

```python
p = sym.plotting.plot3d(f,(x, -8,8),(y, -8,8), title='$f(x,y)=%s$' %sym.latex(f))
p = sym.plotting.plot3d(f,(x, -8,8),(y, -8,8), title='$f(x,y)=%s$' %sym.latex(sym.diff(f,x)))
p = sym.plotting.plot3d(f,(x, -8,8),(y, -8,8), title='$f(x,y)=%s$' %sym.latex(sym.diff(f,y)))
```
